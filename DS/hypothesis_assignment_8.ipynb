```python
import numpy as np
import pandas as pd
from scipy import stats

obs = np.array([
    [50, 70],
    [80,100],
    [60, 90],
    [30, 50],
    [20, 50]
])

row_names = ["Very Satisfied", "Satisfied", "Neutral", "Unsatisfied", "Very Unsatisfied"]
col_names = ["Smart Thermostat", "Smart Light"]
obs_df = pd.DataFrame(obs, index=row_names, columns=col_names)
obs_df
```

```python
row_totals = obs.sum(axis=1)      
col_totals = obs.sum(axis=0)       
grand_total = obs.sum()           

expected = np.outer(row_totals, col_totals) / grand_total

expected_df = pd.DataFrame(expected, index=row_names, columns=col_names)

row_totals, col_totals, grand_total, expected_df
```

```python
chi_square_components = (obs - expected)**2 / expected
chi_square_stat = chi_square_components.sum()

chi_square_components_df = pd.DataFrame(chi_square_components, index=row_names, columns=col_names)
chi_square_stat, chi_square_components_df

```

```python
chi2, p_value, dof, expected_scipy = stats.chi2_contingency(obs, correction=False)

chi2, p_value, dof, expected_scipy
```

```python
alpha = 0.05
df = dof
critical_value = stats.chi2.ppf(1 - alpha, df)

print("Chi-square statistic:", chi2)
print("Degrees of freedom:", df)
print("Critical value at alpha=0.05:", critical_value)
print("p-value:", p_value)

if chi2 > critical_value:
    print("Decision: Reject H0 (there is an association).")
else:
    print("Decision: Fail to reject H0 (no significant association).")
```

 ##### Hypothesis Testing


```python
import numpy as np
from math import sqrt
from scipy import stats

x_bar = 3050.0        
X_mean = 600.0        
sigma_X = 25.0         
n = 25                

mu_W = 1000 + 5 * X_mean    
sigma_W = 5 * sigma_X        

x_bar, mu_W, sigma_W, n
```

```python
se = sigma_W / sqrt(n)              
z_stat = (x_bar - mu_W) / se        

se, z_stat
```

```python
alpha = 0.05
z_critical = stats.norm.ppf(1 - alpha)   

p_value = 1 - stats.norm.cdf(z_stat)

z_critical, p_value
```

```python
print("Sample mean (x̄):", x_bar)
print("Theoretical mean (μ0):", mu_W)
print("Population sd (σ):", sigma_W)
print("SE:", se)
print("Z statistic:", z_stat)
print("Z critical (alpha=0.05, one-tailed right):", z_critical)
print("p-value (right-tailed):", p_value)

if z_stat > z_critical:
    print("Decision: Reject H0 → evidence that mean > 4000 (costs higher).")
else:
    print("Decision: Fail to reject H0 → no evidence that mean > 4000.")
```
