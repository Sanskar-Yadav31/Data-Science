```python
sample = [1.13, 1.55, 1.43, 0.92, 1.25, 1.36, 1.32, 0.85, 1.07, 1.48, 1.20, 1.33, 1.18, 1.22, 1.29]

import statistics

n = len(sample)
mean_x = statistics.mean(sample)
s = statistics.stdev(sample)  

print('Sample size n =', n)
print('Sample mean (x̄) =', round(mean_x, 6))
print('Sample standard deviation (s) =', round(s, 6))
```

```python
from mpmath import mp
import scipy.stats as st

alpha = 0.01
df = n - 1
t_crit = st.t.ppf(1 - alpha/2, df)
SE_s = s / math.sqrt(n)
margin_t = t_crit * SE_s
ci_t_lower = mean_x - margin_t
ci_t_upper = mean_x + margin_t

print('Degrees of freedom (df) =', df)
print('t critical (two-sided, 99%) =', round(t_crit, 6))
print('Standard error (using s) =', round(SE_s, 6))
print('Margin = t_crit * SE =', round(margin_t, 6))
print('\n99% CI using t-distribution: (', round(ci_t_lower, 6), ',', round(ci_t_upper, 6), ')')

```

```python
sigma = 0.2
z_crit = st.norm.ppf(1 - alpha/2)
SE_sigma = sigma / math.sqrt(n)
margin_z = z_crit * SE_sigma
ci_z_lower = mean_x - margin_z
ci_z_upper = mean_x + margin_z

print('z critical (two-sided, 99%) =', round(z_crit, 6))
print('Standard error (using known sigma) =', round(SE_sigma, 6))
print('Margin = z_crit * SE =', round(margin_z, 6))
print('\n99% CI using z-distribution (sigma known = 0.2): (', round(ci_z_lower, 6), ',', round(ci_z_upper, 6), ')')
```

```python
width_t = ci_t_upper - ci_t_lower
width_z = ci_z_upper - ci_z_lower
print('Width of t-based CI =', round(width_t, 6))
print('Width of z-based CI =', round(width_z, 6))

print('\nInterpretation:')
print('- t-based interval is wider because we estimated σ by s (extra uncertainty).')
print('- z-based interval is narrower because σ is assumed known = 0.2.')
```
