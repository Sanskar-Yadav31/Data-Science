```python
import pandas as pd
from itertools import combinations
from collections import Counter
```

```python
# Load the dataset
df = pd.read_excel("D:\\DS Assignment\\Association Rules\\Association Rules\\Online retail.xlsx")

# Display first rows
df.head()
```

```python
# Convert dataset into transaction list

col = df.columns[0]

transactions = df[col].astype(str).apply(
    lambda x: [item.strip().lower() for item in x.split(',') if item.strip()!='']
).tolist()

print("Total Transactions:", len(transactions))
```

```python
min_support = 0.01
min_count = int(min_support * len(transactions))

item_counts = Counter()

for t in transactions:
    item_counts.update(set(t))

frequent_items = {item for item, count in item_counts.items() if count >= min_count}

print("Frequent 1-itemsets:", len(frequent_items))
```

```python
pair_counts = Counter()

for t in transactions:
    for pair in combinations(set(t), 2):
        pair_counts[pair] += 1

frequent_pairs = {pair:count for pair,count in pair_counts.items() if count >= min_count}

print("Frequent 2-itemsets:", len(frequent_pairs))
```

```python
rules = []

for (item1, item2), count in frequent_pairs.items():
    
    support = count / len(transactions)
    
    support_item1 = item_counts[item1] / len(transactions)
    support_item2 = item_counts[item2] / len(transactions)
    
    confidence = support / support_item1
    
    lift = confidence / support_item2
    
    if confidence >= 0.4:
        rules.append([item1, item2, support, confidence, lift])
```

```python
rules_df = pd.DataFrame(
    rules,
    columns=["Item A", "Item B", "Support", "Confidence", "Lift"]
)

rules_df.sort_values(by="Lift", ascending=False).head(10)
```

```python
rules_df.to_csv("association_rules_output.csv", index=False)
```
